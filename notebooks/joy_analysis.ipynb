```python
import pandas as pd

df = pd.read_csv("../data/wrime.tsv", sep="\t")
sentiment = "Joy"
df = df[
    [
        "Sentence",
        f"Writer_{sentiment}",
        f"Reader1_{sentiment}",
        f"Reader2_{sentiment}",
        f"Reader3_{sentiment}",
    ]
].copy()
df
```

```python
reader_columns = [
    f"Reader1_{sentiment}",
    f"Reader2_{sentiment}",
    f"Reader3_{sentiment}",
]
```

```python
df["Std. Readers"] = df[reader_columns].std(axis=1)
```

```python
df.sort_values("Std. Readers").head(10)
```

```python
df.sort_values("Std. Readers").tail(10)
```

```python
df["Std. Readers"].round(6).value_counts().sort_index()
```

```python
df = df[df["Std. Readers"] < 1].copy()  # 客観3人の標準偏差が1未満の文章を抽出
df
```

```python
df.drop("Std. Readers", axis=1, inplace=True)
df["Avg. Readers"] = df[reader_columns].mean(axis=1)
df["|Writer - Avg. Readers|"] = abs(df[f"Writer_{sentiment}"] - df["Avg. Readers"])
```

```python
df.sort_values("|Writer - Avg. Readers|").head(10)
```

```python
df.sort_values("|Writer - Avg. Readers|").tail(10)
```

```python
df[df[f"Writer_{sentiment}"] < df["Avg. Readers"]].sort_values(
    "|Writer - Avg. Readers|"
).tail(10)
```
